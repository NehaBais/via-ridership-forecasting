```python
import matplotlib.pyplot as plt
import seaborn as sns
import polars as pl
import os
```

```python
CLEAN_DATA_FOLDER = "clean_data"
```

```python
ridership = pl.scan_csv(os.path.join(CLEAN_DATA_FOLDER, "ridership.csv"))
```

```python
# randomly sample 50% of the data
# ridership50 = ridership.with_row_index().filter(pl.col("index").hash(42) % 2 == 1).drop("index").collect()
```

```python
_, ax = plt.subplots(1, 1, figsize=(24, 20))
_ = sns.heatmap(
    ridership.collect().corr(),
    vmin=-1,
    vmax=1,
    annot=True,
    center=0,
    square=True,
    xticklabels=ridership.columns,
    yticklabels=ridership.columns,
    ax=ax,
)
```
